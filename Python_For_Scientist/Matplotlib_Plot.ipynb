# Plot Two-Dimensional (2D) lines/curves using Matplotlib

We can use Matplotlib to visualize a function with one variable, e,g. y = f(x)

```python
%matplotlib inline
#%matplotlib notebook
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

First, we create a figure (fig) with axes (ax) using ``plt.subplots``

```python
fig, ax = plt.subplots()
```

The *figure* (fig from plt.subplots()) is a container that holds the objects representing axes, graphics, text, and labels. <br>
The *axes* (ax from plt.subplots()) gives us a coordinate system: a bounding box with horizontal and vertical lines <br>
Once we have created a *fig* with *ax*, we can use the ``ax.plot`` function to plot some data

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
y = np.sin(x)
ax.plot(x, y)
```

We can add two lables 'x' and 'y' to the plot

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
y = np.sin(x)
ax.plot(x, y)
ax.set_xlabel('x', fontsize=20) # change x to x [time], fontsize to 30
ax.set_ylabel('y', fontsize=20) # change y to 'y [amplitude]'
```

If we want to create a single figure with multiple lines, we can simply call the ``plot`` function multiple times:

We can plot multiple curves together

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
ax.plot(x, y1, x, y2)
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y1 and y2', fontsize=20)
```

## Set Line Colors in a Plot

In the above example, we plot a sin curve, and a cos curve <br>
It is not obvious to differentiate between the two curves <br>
For a better visualization, we need two different colors for the two curves.  <br>
We can also change  the style of a curve (which is called line in a plot)

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
#
y0 = np.sin(x)
y1 = np.sin(x+1)
y2 = np.sin(x+2)
y3 = np.sin(x+3)
#
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y0, y1, y2, y3', fontsize=20)
#
ax.plot(x, y0, color='blue')
#ax.plot(x, y0, 'b')
ax.plot(x, y1, color='green')
#ax.plot(x, y1, 'g')
ax.plot(x, y2, color='red')
#ax.plot(x, y2, 'r')
ax.plot(x, y3, color=(1, 1, 0)) # RGB color (red, green blue), the value is from 0 to 1
```

If no color is specified for a curve, Matplotlib will assign a color randomly

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
y0 = np.sin(x)
y1 = np.sin(x+1)
y2 = np.sin(x+2)
y3 = np.sin(x+3)
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y0, y1, y2', fontsize=20)
# one line of code to plot the four curves
ax.plot(x, y0, 'b', x, y1, 'g', x, y2, 'r', x, y3, 'y')
```

## Set Line Styles in a Plot

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
#
y0 = np.sin(x)
y1 = np.sin(x+1)
y2 = np.sin(x+2)
y3 = np.sin(x+3)
#
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y0, y1, y2, y3', fontsize=20)
#
ax.plot(x, y0, color='blue', linestyle='solid')
#ax.plot(x, y0, 'b-')
ax.plot(x, y1, color='green', linestyle='dashed')
#ax.plot(x, y1, 'g--')
ax.plot(x, y2, color='red', linestyle='dashdot')
#ax.plot(x, y2, 'r-.')
ax.plot(x, y3, color='yellow', linestyle='dotted')
#ax.plot(x, y3, 'y:')
```

we can plot four curves in one line of code using different colors and styles

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
y0 = np.sin(x)
y1 = np.sin(x+1)
y2 = np.sin(x+2)
y3 = np.sin(x+3)
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y0, y1, y2, y3', fontsize=20)
# plot four curves in one line of code using different colors and styles
ax.plot(x, y0, 'b-', x, y1, 'g--', x, y2, 'r-.', x, y3, 'y:')
```

## Set Axes Limits of a Plot

We can set the ranges of the horizontal (x) axis and the vertical (y) axis <br>
using the functions ``ax.set_xlim()`` and ``ax.set_ylim()``

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
y = np.sin(x)
ax.plot(x, y)
#
ax.set_xlim(-1, 10)
ax.set_ylim(-10, 10)
#one line of code to set the two ranges
#ax.axis([-1, 10, -10, 10])
```

We can adjust the aspect ratio between the two axes using the function ``ax.set_aspect``

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
y = np.sin(x)
ax.plot(x, y)
ax.axis([-1, 10, -10, 10])
ax.set_aspect(1) # change to other numbers 
#ax.set_aspect('equal')
```

We can get more information about axes in the online document <br>
https://matplotlib.org/api/axes_api.html

## Add a legend (caption) in a Plot

We can add a legend in a plot to show the name of a curve

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
y0 = np.sin(x)
y1 = np.sin(x+1)
y2 = np.sin(x+2)
y3 = np.sin(x+3)
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y0, y1, y2, y3', fontsize=20)
ax.plot(x, y0, color='blue',   linestyle='solid',   label = 'y0=sin(x)') # set the label
ax.plot(x, y1, color='green',  linestyle='dashed',  label = 'y1=sin(x+1)')
ax.plot(x, y2, color='red',    linestyle='dashdot', label = 'y2=sin(x+2)')
ax.plot(x, y3, color='yellow', linestyle='dotted',  label = 'y3=sin(x+3)')
ax.axis([-1, 20, -1, 1])
#show the legend
ax.legend(loc='upper right', fontsize=16)
```

## Add a Title to a Plot

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
y0 = np.sin(x)
y1 = np.sin(x+1)
y2 = np.sin(x+2)
y3 = np.sin(x+3)
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y0, y1, y2, y3', fontsize=20)
ax.plot(x, y0, color='blue',   linestyle='solid',   label = 'y0=sin(x)') # set the label
ax.plot(x, y1, color='green',  linestyle='dashed',  label = 'y1=sin(x+1)')
ax.plot(x, y2, color='red',    linestyle='dashdot', label = 'y2=sin(x+2)')
ax.plot(x, y3, color='yellow', linestyle='dotted',  label = 'y3=sin(x+3)')
ax.axis([-1, 20, -1, 1])
ax.legend(loc='upper right', fontsize=16)
#add a tile
ax.set_title('Four sin curves in one plot', fontsize=16)
```

## Show four curves in four axes of a figure

```python
#create a figure with four axes
fig, ax = plt.subplots(2, 2, constrained_layout=True)
```

```python
ax.shape
```

ax is a 2D numpy array. It has four elements/axes: ax[0,0],  ax[0,1],  ax[0,2],  ax[0,3]

```python
#draw the curve y0 = np.sin(x) in axes[0,0]
fig, ax = plt.subplots(2, 2, constrained_layout=True)
x = np.linspace(0, 10, 1000)
y0 = np.sin(x)
#---------------------------------------------
ax[0,0].set_xlabel('x', fontsize=16)
ax[0,0].set_ylabel('y0', fontsize=16)
ax[0,0].plot(x, y0, color='blue',   linestyle='solid',   label = 'y0=sin(x)')
ax[0,0].axis([-1, 10, -1, 1])
ax[0,0].set_title('y0=sin(x)', fontsize=16)
# add a text
ax[0,0].text(0, -0.9, 'ax[0,0]', fontsize=16)
```

```python
#draw the four curves
fig, ax = plt.subplots(2, 2, constrained_layout=True)
x = np.linspace(0, 10, 1000)
y0 = np.sin(x)
y1 = np.sin(x+1)
y2 = np.sin(x+2)
y3 = np.sin(x+3)
#---------------------------------------------
ax[0,0].set_xlabel('x', fontsize=16)
ax[0,0].set_ylabel('y0', fontsize=16)
ax[0,0].plot(x, y0, color='blue',   linestyle='solid')
ax[0,0].set_title('y0=sin(x)', fontsize=16)
ax[0,0].text(0, -0.9, 'ax[0,0]', fontsize=16)
#---------------------------------------------
ax[0,1].set_xlabel('x', fontsize=16)
ax[0,1].set_ylabel('y1', fontsize=16)
ax[0,1].plot(x, y1, color='green',   linestyle='solid')
ax[0,1].set_title('y1=sin(x+1)', fontsize=16)
ax[0,1].text(0, -0.9, 'ax[0,1]', fontsize=16)
#---------------------------------------------
ax[1,0].set_xlabel('x', fontsize=16)
ax[1,0].set_ylabel('y2', fontsize=16)
ax[1,0].plot(x, y2, color='red',   linestyle='solid')
ax[1,0].set_title('y2=sin(x+2)', fontsize=16)
ax[1,0].text(0, -0.9, 'ax[1,0]', fontsize=16)
#---------------------------------------------
ax[1,1].set_xlabel('x', fontsize=16)
ax[1,1].set_ylabel('y0', fontsize=16)
ax[1,1].plot(x, y3, color='yellow',   linestyle='solid')
ax[1,1].set_title('y3=sin(x+3)', fontsize=16)
ax[1,1].text(0, -0.9, 'ax[1,1]', fontsize=16)
```

# Three-Dimensional (3D) Plot in Matplotlib
Matplotlib was initially designed for 2D visualization. <br>
After the release of v1.0, Matplotlib has a limited number of functions for 3D visualization in the module mplot3d of mpl_toolkits

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
# the argument "projection='3d'" is needed to create a 3D axes
```

## Plot 3D Points and Lines

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

# point data for a 3D curve
z = np.linspace(0, 10, 100)
x = np.sin(z)
y = np.cos(z)
ax.plot(x, y, z, 'r-') # change '.' to '-' to show the curve
ax.set_xlabel('X', fontsize=16)
ax.set_ylabel('Y', fontsize=16)
ax.set_zlabel('Z', fontsize=16)
ax.set_title('a trajectory of an object in a 3D space', fontsize=16)
```

## Plot 3D  Surfaces

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap='viridis', linewidth=0, antialiased=False)
```
